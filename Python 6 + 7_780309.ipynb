```python
# group by
# .groupby('')
# .groupby(['', ''])

# .groupby('')[['']] or .groupby('')[['', '']]
# .groupby('').mean() / some other aggregate

# columns mentioned in group by would be now used as index or row labels
# if no selection of columns is mentioned then said aggregation will be applied on each column of raw df except for grouped cols

# multiple aggregations in the same result
# .agg
# df[''].agg(['min', 'max']) -> 2 different values -> 1 for min and 1 for max
# .groupby('')[['']].agg(['max', 'mean', 'sum']) -> 3 cols (1 each for each agg) / what about the row labels? - group by cols

# .first()
# .last()
# .nth() -> index value assigned to rows after sorting basis desc order
```
