## splitting..

```python
X_train = tr_df.drop(['booking_status','Date','Booking_ID'],axis=1)
X_test = tt_df.drop(['booking_status','Date','Booking_ID'],axis=1)
```

```python
y_train = tr_df['booking_status']
y_test = tt_df['booking_status']
```

## logisitc regression

```python
model = LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
clf = classification_report(y_test,y_pred)
print(clf)
```

```python
#training accuracy
accuracy_score(y_train,model.predict(X_train))
```

```python
#testing accuracy
accuracy_score(y_test,model.predict(X_test))
```

## Decision tree

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
```

```python
clf = classification_report(y_test,y_pred)
print(clf)
#training accuracy
print(accuracy_score(y_train,model.predict(X_train)))
#testing accuracy
print(accuracy_score(y_test,model.predict(X_test)))
```

## Bagging..

```python
clf = BaggingClassifier(DecisionTreeClassifier(),n_estimators=100,max_samples=12,random_state=42)
clf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
clfrprt = classification_report(y_test,y_pred)
print(clfrprt)
#training accuracy
print(accuracy_score(y_train,clf.predict(X_train)))
#testing accuracy
print(accuracy_score(y_test,clf.predict(X_test)))
```

## Boosting..

```python
booster = AdaBoostClassifier(DecisionTreeClassifier(max_depth=5),n_estimators=3,random_state=42)
booster.fit(X_train,y_train)
y_pred = booster.predict(X_test)
```

```python
clf = classification_report(y_test,y_pred)
print(clf)
#training accuracy
print(accuracy_score(y_train,booster.predict(X_train)))
#testing accuracy
print(accuracy_score(y_test,booster.predict(X_test)))
```

### Boosting gives the highest test accuracy
