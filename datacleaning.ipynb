```python
data = {'Name': ['A', 'B', 'C', 'D', 'E', 'F'],
       'Score': [30, 40, 10, 50, 40, 40 ]}
```

```python
df3 = pd.DataFrame(data)
df3
```

```python
# what ranks would be assigned to person B, E and F?
```

```python
df3['rank'] = df3['Score'].rank()

df3
```

```python
# ideal rank for person B if he does not tie with E and F and his score is less than that of E and F? -> 3
# E -> 4
# F -> 5
(3 + 4 + 5)
```

```python
# method = 'min'

df3['min_rank'] = df3['Score'].rank(method = 'min')
df3

# similar to rank() in SQL
```

```python
# method = 'max'

df3['max_rank'] = df3['Score'].rank(method = 'max')

df3
```

```python
# method = 'dense'

df3['D_rank'] = df3['Score'].rank(method = 'dense')
```

```python
df3
```

```python
# rank() skips ranks when tied values are encountered
# next rank after tied rank assignment? -> tied rank + num of rows having the tied rank = 3 + 3 = 6
# dense-rank() -> tied scores get the same rank, next rank is just an increment of 1 on tied rank

```

```python
# method = 'first'

df3['row_num'] = df3['Score'].rank(method = 'first')

df3
```

```python
# row_number() works like serial number
# So it does not allow for repeated values.
```

```python
# 1. default (average) -> assigns average of ranks when tied values are encountered
# 2. method = 'min' -> assign the least rank to all repeated values (similar to rank() in SQL)
# 3. method = 'max' -> assign the highest rank to all repeated values amongst the repeated values (similar to rank() in SQL)
# 4. method = 'dense' -> no skipping of ranks after assigning ranks to tied values (similar to dense_rank() in SQL)
# 5. method = 'first' -> similar to row_num in SQL
```

```python
data = {'Name': ['A', 'B', 'C', 'D', 'E'],
    'Department': ['IT', 'HR', 'IT', 'IT','HR'],
       'Salary': [100, 100, 200,150, 300]}

df4 = pd.DataFrame(data)

df4
```

```python
# rank employees amongst their department members basis salary (asc)

df4['company_rank'] = df4['Salary'].rank(method = 'dense')
df4
```

```python
# groupby
# group by basis which column? -> 
```

```python
df4.groupby('Department')['Salary'].rank()
```

```python
df4['dept_rank'] = df4.groupby('Department')['Salary'].rank()
df4
```

```python
df4.sort_values(by = 'Department')

# ORDER BY Department -> In SQL
```

```python
# secondary sort the above result basis asc order of dept_rank
# ORDER BY Department, Dept_rank

df4.sort_values(by = ['Department', 'dept_rank'])
```

```python
# secondary sort the above result basis asc order of dept_rank
# ORDER BY Department, Dept_rank DESC

df4.sort_values(by = ['Department', 'dept_rank'], ascending = False)
```

```python
# secondary sort the above result basis asc order of dept_rank
# ORDER BY Department, Dept_rank DESC

df4.sort_values(by = ['Department', 'dept_rank'], ascending = [True, False])
```

```python
# when to use rank
# IIT B -> CS 
# JEE AD
# IITB -> CS -> 60 SEATS TO FILL IN WITH INDIVIDUAL STUDENTS
# IITB -> TOP 60 in the rank list
# rank, dense_rank
# 300 TOTAL MARKS -> rank = 1
# 299 -> 2
# Many students can get the same marks
# rank -> starts with 1 and ends with len of table / num of rows
# dense_rank -> starts with 1 and end rank is not determinable unless data is known
# when using dense_rank by the time you get to rank 60, there can be more than 60 students with ranks less than or equal to 60
# rank -> number of students by the time you get to rank = 60 is going to 60 or less but never more than 60
```
