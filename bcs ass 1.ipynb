#### Task A: Identify the star product for the company?

```python
df.describe()
```

```python
prod_quant = pd.DataFrame(df.groupby('item_name')['quantity'].sum())
prod_quant = prod_quant.reset_index().sort_values(by = 'quantity', ascending=False)
prod_quant = prod_quant.reset_index(drop=True)
prod_quant
# 
```

```python
prod_quant['row_num'] = prod_quant.quantity.rank(method='first',ascending=False)
prod_quant
```

```python
star_product = prod_quant[prod_quant.row_num == 1.0]
star_product
```
