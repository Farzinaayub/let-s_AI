```python
# defining dataframes

a = pd.DataFrame(np.random.randint(1, 9, (4, 3)), columns = ['A', 'B', 'c'])

a
```

```python
# Apply inner join on a and b dataframes
# Question you ask: What is the joining condition?

# SELECT * FROM Table_A as A 
# JOIN Table_B as B
# ON A.col_A = B.col_A
```

```python
# Concat
# Concat or combine multiple dataframes along a particular axis 
# how many axes do you deal with when it comes to a dataframe?
# 2 axes - rows and columns or x and y
# vertical concat - axis = 0
# horizontal concat - axis = 1
```

```python
a = pd.DataFrame({'A': [1,2,3], 'B': ['a', 'b', 'c']})

a
```

```python
b = pd.DataFrame({'A' : [4,5,6], 'B': ['d', 'e', 'f']})
b
```

```python
# default axis = 0
# axis = 0, vertical concat
# dataframes being placed one above the other
# top to bottom stacking of dataframes

pd.concat([a,b])
```

```python
# axis = 1
# left to right stacking of dataframes

pd.concat([a,b] , axis = 1)
```

```python
a
```

```python
b
```

```python
# restart indexing 
# ignore previous values

pd.concat([a,b], ignore_index = True)
```
