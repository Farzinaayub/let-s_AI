```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn import metrics
import seaborn as sns
```

```python
df=pd.read_csv('datasets\Housing Habitability Data.csv')
df.head()
```

```python
df.Number_of_Windows.fillna(0,inplace=True)
```

```python
df.Number_of_Windows=df.Number_of_Windows.astype(int)
```

```python
df.Furnishing.fillna('Unfurnished',inplace=True)
```

```python
df['Frequency_of_Powercuts'].mode()
```

```python
df.Frequency_of_Powercuts.fillna(0,inplace=True)
```

```python
df.Frequency_of_Powercuts=df.Frequency_of_Powercuts.astype(int)
```

```python
df['Crime_Rate'].mode()
```

```python
df.Crime_Rate.fillna('Well below average',inplace=True)
```

```python
df['Dust_and_Noise'].mode()
```

```python
df.Dust_and_Noise.fillna('Medium',inplace=True)
```

```python
desc=pd.DataFrame(index=list(df))
desc['count']=df.shape[0]
desc['null']=df.isnull().sum()
desc['dtype']=df.dtypes
desc['unique']= df.nunique()
desc['%unique'] = desc['unique']/len(df)
desc=pd.concat([desc,df.describe().T],axis=1)
```

```python
x = df[['Property_Area','Number_of_Windows','Number_of_Doors','Frequency_of_Powercuts','Traffic_Density_Score','Air_Quality_Index','Neighborhood_Review']]
y = df['Habitability_score']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```
